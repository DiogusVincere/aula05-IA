```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Criando o conjunto de dados
vazao = np.array([10,20,30,5,35,40]).reshape(-1,1) # Cria o conjunto de dados
nivel = np.array([50,60,70,45,80,85]) # conjunto de dados para o nivel
# Cria modelo e treina o modelo de regressão linear
modelo = LinearRegression() # Armazena a regressão linear na variavel modelo
modelo.fit(vazao, nivel) # Treina o modelo de regressão linear
```

```python
# Teste do modelo criado
vazao_teste = np.array([100,50,40,30,20,10,0]).reshape(-1,1) # Vazão de água em m3/min
#vazao_teste = np.array([5,2,7,30,50]).reshape(-1,1)
# Fazer previsões com base na vazão de água
nivel_previsto = modelo.predict(vazao_teste) # Realiza a previsão de nivel do reserva

# Imprimir as previsões 
for i in range(len(vazao_teste)):
    print(f'Vazão: {vazao_teste[i][0]} m3/min - Nivel previsto: {nivel_previsto[i]} metros')
plt.scatter(vazao,nivel,color='blue',label='Dados do Treinamento')
plt.plot(vazao_teste,nivel_previsto,color='red',label='Dados de teste')
plt.xlabel('Vazão de Água (m3/min)')
plt.ylabel('Nível do Tanque (metros)')
plt.title('Relação entre Vazão de Água e Nível do Tanque')
plt.legend() # Exibe a legenda do gráfico
plt.grid(True) # Exibe a grade do gráfico
plt.show() # Plota o Gráfico
```

```python
Quantidade = np.array([100,200,300,400,500,600,700]).reshape(-1,1)
Temperatura = np.array([30,35,40,45,60,100,150])

# Cria modelo e treina o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(Quantidade,Temperatura)
```

```python
# Teste do modelo criado
Quantidade_teste = np.array([95,170,290,385,450,565,690]).reshape(-1,1)
#Quantidade_teste = np.array([5,2,7,30,50]).reshape(-1,1)
Temperatura_previsto = modelo.predict(Quantidade_teste)

# Imprimir as previsões 
for i in range(len(Quantidade_teste)):
    print(f'Calor: {Quantidade_teste[i][0]} - Temperatura previsto: {Temperatura_previsto[i]}')
plt.scatter(Quantidade,Temperatura,color='blue',label='Dados do Treinamento')
plt.plot(Quantidade_teste,Temperatura_previsto,color='red',label='Dados de teste')
plt.xlabel('Quandidade de Calor')
plt.ylabel('Temperatura do Forno')
plt.title('Relação entre Quantidade de Calor e Temperatura do forno')
plt.legend() # Exibe a legenda do gráfico
plt.grid(True) # Exibe a grade do gráfico
plt.show() # Plota o Gráfico
```

```python
area = np.array([50,100,120,300]).reshape(-1,1)
valor = np.array([180.000,300.000,375.000,600.000])

# Cria modelo e treina o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(area,valor)
```

```python
# Teste do modelo criado
area_teste = np.array([45,95,105,270]).reshape(-1,1)
#Quantidade_teste = np.array([5,2,7,30,50]).reshape(-1,1)
valor_previsto = modelo.predict(area_teste)

# Imprimir as previsões 
for i in range(len(area_teste)):
    print(f'Área: {area_teste[i][0]} - Valor previsto: {valor_previsto[i]}')
plt.scatter(area,valor,color='blue',label='Dados do Treinamento')
plt.plot(area_teste,valor_previsto,color='red',label='Dados de teste')
plt.xlabel('Área do imovel')
plt.ylabel('Valor do imovel')
plt.title('Relação entre Área e VAlor do imovel')
plt.legend() # Exibe a legenda do gráfico
plt.grid(True) # Exibe a grade do gráfico
plt.show() # Plota o Gráfico
```

```python
#cria o dicionario com o estoque de cada loja
dic_1 = {
    'Lojas':['A','B','C','D'],
    'Estoque':[500,100,300,600],
    'Chave':['AA','BB','CC','DD']
}
df_estoque = pd.DataFrame(dic_1)
```

```python
df_estoque.head()
```

```python
dic_vendas={
    'Lojas':['A','B','C','D'],   
    'Jan':[100,80,150,50],
    'Fev':[120,90,170,60],
    'Mar':[150,100,200,80],
    'Abril':[180,110,230,90],
    'Mai':[220,190,350,200],
    'Jun':[230,150,2980,120],
    'Jul':[150,170,300,140],
    'Ago':[260,180,310,150],
    'Set':[240,160,290,130],
    'Out':[220,140,270,110],
    'Nov':[400,200,350,190],
    'Dez':[300,350,400,250],
    'Chave':['AA','BB','CC','DD']

}
df_vendas = pd.DataFrame(dic_vendas)
```

```python
df_vendas.head()
```

```python
df_ecommerce = pd.merge(df_estoque,df_vendas,how='inner',on='Chave')
```

```python
df_ecommerce.head()
```

```python
df_ecommerce.drop('Lojas_y',axis=1,inplace=True) # Exclui a coluna Lojas_y
```

```python
df_ecommerce.head()
```

```python
df_ecommerce.drop('Chave',axis=1,inplace=True)
```

```python
df_ecommerce.head()
```

```python
df_ecommerce = df_ecommerce.T
```

```python
df_ecommerce.head()
```

```python
df_ecommerce.set_index(['Lojas_x'])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5)) # Cria uma figura e armazena
# axs - ax - é a abreviação de axes
# S é de subplot

# Subplot 1: Gráfico de barras
axs[0].bar(df_ecommerce['Lojas_x'], df_ecommerce['Estoque'])
axs[0].set_title('Estoque de produtos')
axs[0].set_xlabel('Lojas')
axs[0].set_ylabel('Quantidade de produtos')

axs[0].grid()
# Subplot 2: Boxplot
axs[1].bar(df_ecommerce['Lojas_x'], df_ecommerce['Jan'])
axs[1].set_title('Histórico de vendas')
axs[1].set_xlabel('Lojas')
axs[1].set_ylabel('Quantidade de produtos')
```

```python
# utliza a biblioteca seaborn para plotar o gráfico de barras para
# cada loja com a quantidade de vendas no mes de fevereiro
sns.barplot(x='Fev', y='Lojas_x', data=df_ecommerce, orient='h')
plt.title('Vendas em Fevereiro')
plt.show()
```

```python
# Dados de vendas
dic_vendas = {
    'Lojas': ['A', 'B', 'C', 'D'],
    
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abril': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 290, 120],
    'Jul': [150, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 200, 350, 190],
    'Dez': [300, 350, 400, 250],
    'Chave': ['AA', 'BB', 'CC', 'DD']
    
}
df_vendas = pd.DataFrame(dic_vendas)

# Definindo os meses para o eixo x
meses = ['Jan', 'Fev', 'Mar', 'Abril', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out',
          'Nov', 'Dez']

# Configurando o gráfico
plt.figure(figsize=(10, 6))

# Iterando por cada loja para plotar seu histórico de vendas
for i, loja in enumerate(df_vendas['Lojas']):
    plt.plot(meses, df_vendas.iloc[i, 1:-1], label=f'Loja {loja}', marker='o')

# Adicionando títulos e rótulos
plt.title('Histórico de Vendas de Janeiro a Dezembro')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.legend(title="Lojas")
plt.grid(True)

# Exibindo o gráfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
